### Задача 1 (Парадокс инспектора)

Автобусы отходят от автостанции с 8:00 до 20:00. Первый автобус отходит ровно в 8:00. Затем интервалы между автобусами случайны, независимы и равновероятно равны либо 5-и, либо 10-и минутам. Будем считать, что за минуту на автостанцию приходит ровно один пассажир, и все пассажиры едут ближайшим автобусом.

Проведите $10^4$ экспериментов и с их помощью:

__а) [5 баллов]__ Постройте гистограмму количества автобусов, отошедших от автостанции за сутки. Похоже ли визуально распределение на нормальное?

__б) [10 баллов]__ Инспектор Тимон выбирает равновероятно один из всех автобусов отошедших от автостанции за сутки. Постройте гистограмму числа пассажиров на этом автобусе. Оцените математическое ожидание и дисперсию этого числа.

__в) [10 баллов]__ Инспектор Пумба приходит на автостанцию в случайный момент времени, равномерный от 8:00 до 20:00 и садится в первый пришедший автобус. Постройте гистограмму числа пассажиров на этом автобусе. Оцените математическое ожидание и дисперсию этого числа.

__г) [10 баллов]__ Как изменятся ответы на эти вопросы, если время между автобусами будет экспоненциально со средним в 10 минут?

__Пояснения:__ первый автобус можно не включать в выборку, Пумбу за пассажира можно не считать. 

```python
n = 10 ** 4

counter_of_buses = [] # подсчет количества автобусов за сутки
counter_of_pass = [] # для задания б: количество пассажиров в выбранном Тимоном автобусе 
counter_of_pass_uniform = [] # для задания в: количество пассажиров в выбранном Пумбой автобусе 
for _ in range(n):
    time_start = 8 * 60
    time_finish = 20 * 60
    
    count_bus = 0
    count_pass = []
    flag = 0
    time_Pumba = np.random.uniform(low=time_start, high=time_finish) # равномерное распределение для времени, когда пришел Пумба
    while time_start < time_finish:
        time_now = np.random.choice([5, 10], size=1, p=[0.5, 0.5])[0] # равновероятно выбираем 5 или 10 минут
        time_start += time_now
        count_bus += 1
        count_pass.append(time_now)
        if time_Pumba <= time_start and flag == 0:
          flag = 1
          counter_of_pass_uniform.append(time_now)

    counter_of_pass.append(np.random.choice(count_pass)) # равновероятно

    counter_of_buses.append(count_bus)
```

Пункт а.

```python
plt.figure(figsize=(8, 6))

plt.hist(counter_of_buses, bins=100)
plt.ylabel('Количество экспериментов')
plt.xlabel('Количество автобусов')
plt.title('Количество автобусов, отошедших от автостанции за сутки')
```

Визуально данный график напоминает нормальное распределение, так как мы равновероятно выбирали перерыв между автобусами. Для выбора нам давали всего два варианта (5 или 10).

Рассмотрим возможные варианты: 

Например, получилось так, что выпал вариант 10 минут. Это значит, что для остальных автобусов остаётся меньше времени. Наименьшее количество автобусов мы получим в том случае, если каждый раз выбирать перерыв в 10 минут, но такой вариант маловероятен, поэтому количество таких экспериментов невысоко. 

Также работает и в другую сторону: наибольшее количество автобусов - все перерывы равны 5 минутам. Вероятность этого низка, следовательно экспериментов таких мало.

Наиболее частные варианты, когда перерывы в 5 и 10 минут комбинируются, но и количество автобусов тогда не самое высокое, но и не самое низкое. Также в данном случае количество экспериментов будет большим, так как мы используем равновероятный выбор.

Пункт б.

```python
plt.figure(figsize=(8, 6))

plt.hist(counter_of_pass)
plt.ylabel('Количество экспериментов')
plt.xlabel('Количество пассажиров')
plt.title('Количество пассажиров при равновероятном выборе автобуса')

print('Матожидание = ', np.mean(counter_of_pass))
print('Дисперсия = ', np.var(counter_of_pass))
```

```python
print('Матожидание = ', np.mean(counter_of_pass))
print('Дисперсия = ', np.var(counter_of_pass))
```

Можно заметить, что матожидание, которое мы получили равняется среднему между 5 и 10. Так происходит из-за использования равновероятного выбора между данными числами.

При равновероятном выборе автобуса, мы можем попасть с одинаковой вероятностью в тот, который взял 5 пассажиров (т.е. перерыв был 5 минут), или же в тот, который пришёл после 10-минутного перерыва.

Пункт в.

```python
plt.figure(figsize=(8, 6))

plt.hist(counter_of_pass_uniform)
plt.ylabel('Количество экспериментов')
plt.xlabel('Количество пассажиров')
plt.title('Количество пассажиров при равномерном')

print('Матожидание = ', np.mean(counter_of_pass_uniform))
print('Дисперсия = ', np.var(counter_of_pass_uniform))
```

При равномерном выборе времени, когда приходит Пумба, видно, что вероятность того, что в его автобусе будет 10 пассажиров, выше, так как вероятность, что мы попадем в интервал времени равный 10 минутам, становится выше.

Пункт г.

```python
n = 10 ** 4

counter_of_buses = [] # подсчет количества автобусов за сутки
counter_of_pass = [] # для задания б: количество пассажиров в выбранном Тимоном автобусе 
counter_of_pass_uniform = [] # для задания в: количество пассажиров в выбранном Пумбой автобусе 
for _ in range(n):
    time_start = 8 * 60
    time_finish = 20 * 60
    
    count_bus = 0
    count_pass = []
    flag = 0
    time_Pumba = np.random.uniform(low=time_start, high=time_finish) # равномерное распределение для времени, когда пришел Пумба
    while time_start < time_finish:
        time_now = expon.rvs(scale=10) # экспоненциальное распределение решает, когда поедет следующий автобус
        time_start += time_now
        count_bus += 1
        count_pass.append(int(time_now))
        if time_Pumba <= time_start and flag == 0:
          flag = 1
          counter_of_pass_uniform.append(int(time_now))

    counter_of_pass.append(np.random.choice(count_pass)) # равновероятно

    counter_of_buses.append(count_bus)
```

```python
plt.figure(figsize=(8, 6))

plt.hist(counter_of_buses, bins=100)
plt.ylabel('Количество экспериментов')
plt.xlabel('Количество автобусов')
plt.title('Количество автобусов, отошедших от автостанции за сутки')
```

```python
print('Матожидание = ', np.mean(counter_of_pass))
print('Дисперсия = ', np.var(counter_of_pass))
```

```python
print('Матожидание = ', np.mean(counter_of_pass_uniform))
print('Дисперсия = ', np.var(counter_of_pass_uniform))
```

Можно заметить, что при равновероятном выборе длины перерыва между автобусами график количества автобусов за сутки был более гладким, чем при экспоненциальном распределении данной величины.

Также видно, что при равновероятном выборе перерыва у нас было всего два варианта продолжительности интервала времени (5 или 10 минут), следовательно, и вариантов для количества пассажиров было всего два (так как за одну минуту приходит один пассажир), а при использовании экспоненциального распределения вариантов стало намного больше.

Вследствие разброс значений для количества пассажиров заметно вырос, что видно по росту матожидания и дисперсии.

### Задача 2 (Парадокс Хуана Мануэля Родригеса Паррондо)

У Атоса, Портоса и Арамиса по 1000 франков. Атос постоянно ходит в казино А, где каждый раз выигрывает один франк с вероятностью $0.49$ и проигрывает один франк с вероятностью $0.51$. Портос ходит в казино Б, где ситуация интереснее :) 

- Если богатство посетителя кратно трём, то посетитель выигрывает франк с вероятностью $0.09$ и проигрывает один франк с вероятностью $0.91$
- Если богатство посетителя не кратно трём, то посетитель выигрывает франк с вероятностью $0.74$ и проигрывает один франк с вероятностью $0.26$

Арамис каждый раз выбирает казино А или казино Б равновероятно. Проведите $10^4$ симуляций эволюции благосостояния Атоса, Портоса и Арамиса.Каждая симуляция предусматривает $1000$ посещений казино. Все три игрока все деньги носят с собой и ни на что не тратят :)

__а) [10 баллов]__ Постройте на графике $10$ случайных траекторий изменений богатства Атоса, по горизонтали — номер посещени казино, по вертикали — богатство. На том же графике постройте усреднённую по всем экспериментам тракторию изменения богатства.

__б) [10 баллов]__ Постройте на графике $10$ случайных траекторий изменений богатства Портоса, по горизонтали —  номер посещени казино, по вертикали — богатство. На том же графике постройте усреднённую по всем экспериментам тракторию изменения богатства.

__в) [10 баллов]__ Постройте на графике $10$ случайных траекторий изменений богатства Арамиса, по горизонтали — номер посещени казино, по вертикали — богатство. На том же графике постройте усреднённую по всем экспериментам тракторию изменения богатства.
    

```python
def casino_Portos(w):
    if w % 3 == 0:
        return np.random.choice([-1, 1], size=1, p=[0.91, 0.09])[0]
    return np.random.choice([-1, 1], size=1, p=[0.26, 0.74])[0]

def casino_Atos(w):
    return np.random.choice([-1, 1], size=1, p=[0.51, 0.49])[0]

def casino_Aramis(w):
    casino = np.random.choice([0, 1], size=1, p=[0.5, 0.5])[0]
    if casino == 0:
        return casino_Atos(w) 
    return casino_Portos(w)
```

```python
n = 10 ** 4

Atos, Portos, Aramis = [], [], []

for _ in range(n):
    Atos_w, Portos_w, Aramis_w = 1000, 1000, 1000
    Atos_now, Portos_now, Aramis_now = [], [], []

    for i in range(1000):
        Atos_w += casino_Atos(Atos_w)
        Portos_w += casino_Portos(Portos_w)
        Aramis_w += casino_Aramis(Aramis_w)

        Atos_now.append(Atos_w)
        Portos_now.append(Portos_w)
        Aramis_now.append(Aramis_w)

    Atos.append(Atos_now)
    Portos.append(Portos_now)
    Aramis.append(Aramis_now)
```

Выберем случайным образом индексы для построения графиков.

```python
inds = np.random.choice(np.arange(10 ** 4), size=10)
```

Пункт а.

```python
plt.figure(figsize=(16, 10))

for i in range(10):
    plt.plot(Atos[inds[i]])

plt.plot(np.mean(Atos, axis=0), linewidth=5, color='indigo', label='Средняя траектория')

plt.ylabel('Богатство')
plt.xlabel('Номер посещения казино')
plt.title('Богатство Атоса')
plt.legend()
```

Можно заметить, что большинство из случайно выбранных 10 траекторий богатства Атоса приводят его к уменьшению накоплений. По данной выборке вероятность увеличения баланса всего 1/10.

По усреднённой траектории видно, что тактика Атоса не выигрышная, так как баланс уменьшается в среднем на 20 франков.

Пункт б.

```python
plt.figure(figsize=(16, 10))

for i in range(10):
    plt.plot(Portos[inds[i]])

plt.plot(np.mean(Portos, axis=0), linewidth=5, color='indigo', label='Средняя траектория')

plt.ylabel('Богатство')
plt.xlabel('Номер посещения казино')
plt.title('Богатство Портоса')
plt.legend()
```

У Портоса траектории богатства также не особо позитивны, но они выглядят более стабильными, так как графики можно разбить на отдельные ступени. Так происходит из-за того, что вероятность его проигрыша низкая при кратности баланса 3, в остальных же случаях вероятность победы выше, чем у Атоса.

Однако усреднённая траектория богатства Портоса также показывает уменьшение баланса, но количество потерянных франков немного меньше (по сравнению в Атосом).

Пункт в.

```python
plt.figure(figsize=(16, 10))

for i in range(10):
    plt.plot(Aramis[inds[i]])

plt.plot(np.mean(Aramis, axis=0), linewidth=5, color='indigo', label='Средняя траектория')

plt.ylabel('Богатство')
plt.xlabel('Номер посещения казино')
plt.title('Богатство Арамиса')
plt.legend()
```

Несмотря на то, что две предыдущии тактики оказались проигрышными, их комбинация, которой пользуется Арамис, дала положительные результаты.

Половина из случайно выбранных траекторий приносит ему увеличение накоплений, а также усреднённая траектория для Арамиса даёт ему прибыль примерно равную 10 франкам.

Так происходит из-за того, что за счёт случайного выбора казино Арамис реже, чем Портос, теряет франки при кратности его баланса 3, а также благодаря тому, что в случаях, когда он попадает в казино Портоса при накоплениях не кратным 3, вероятность его выигрыша франка выше, чем у Атоса.

### Задача 3 (Парадокс Берксона)

Предположим, что результаты ЕГЭ школьников по русскому и математике независимы и хорошо аппроксимируются 
нормальным распределением с ожиданием 60 и стандартным отклонением 10.

__а) [5 баллов]__ Постройте график числа прошедших в УШЭ в зависимости от $n$.

__в) [5 баллов]__ Постройте график выборочной корреляции между результатами по русскому и математике в зависимости от $n$ среди не прошедших в УШЭ.

```python
counter_of_students = []
math = []
russian = []

math_not = []
russian_not = []

n_all = np.arange(80, 161, 5)
for n in n_all:
    count = 0

    add_math = []
    add_russian = []

    add_math_not = []
    add_russian_not = []

    for _ in range(10 ** 4):
        first_exam = np.random.normal(loc=60, scale=10)
        second_exam = np.random.normal(loc=60, scale=10)
        if first_exam + second_exam >= n:
            add_math.append(first_exam)
            add_russian.append(second_exam)
            count += 1
        else:
            add_math_not.append(first_exam)
            add_russian_not.append(second_exam)

    math.append(np.array(add_math))
    russian.append(np.array(add_russian))
    math_not.append(np.array(add_math_not))
    russian_not.append(np.array(add_russian_not))
    counter_of_students.append(count)
```

Пункт а.

```python
plt.figure(figsize=(16, 10))

plt.plot(counter_of_students)
plt.xticks(np.arange(17), n_all)

plt.ylabel('Прошедшие студенты')
plt.xlabel('Порог')
plt.title('Зависимость между порогом и количеством студентов')
plt.legend()
```

По графику видно, что при низком пороге почти все студенты проходят, так как минимальный проходной балл равен 80, а используемое распределение имеет среднее равное 60 с стандартным отклонением 10, т.е. в таком случае все выдаваемые значения в сумме будут превосходить данный проходной бал.

При увеличении порога количество прошедших студентов уменьшается, так как мы пользуемся тем же распределением, которое в среднем выдает значение 60. Именно поэтому на графике можно заметить, что при пороге равном 120 линия графика становится вогнутой (меняет свою форму), так как при среднем значении 60, учитывая суммирование баллов по двум экзаменам, мы получаем данное значение. После прохождения порога равного 120 баллам график начинает убывать медленнее. 

Пункт б.

```python
math_ = np.array(math)
russian_ = np.array(russian)

corr = []
for i in range(len(math_)):
    corr_now = np.corrcoef(math_[i], russian_[i])
    corr.append(corr_now[0][1])
```

```python
plt.figure(figsize=(16, 10))

plt.plot(corr)
plt.xticks(np.arange(17), n_all)

plt.ylabel('Значение корреляции')
plt.xlabel('Порог')
plt.title('Зависимость между порогом и корреляцией между результатами за экзамен среди прошедших')
plt.legend()
```

Вспомним: корреляция близка к нулю -> нет линейной зависимости; корреляция по модулю близка к 1 -> есть зависимость.

На графике можно заметить, что при увеличении порога корреляция приближается к -1, что означает, что здесь присутствует обратнопропорциональная зависимость, и с увеличением порога она растёт.

Вспомним, что мы смотрим на корреляцию между баллами за русский и математику у прошедших студентов.

Корреляция близка к 0 в случаях, когда порог минимален. Это объясняется тем, что (как я говорила ранее) при низком пороге довольно легко пройти, т.е. можно получить низкий бал по математике и нормальный по русскому и в итоге набрать 80 баллов -> пройти порог.

Также корреляция стремится к 1 при увеличении порога. Также объясняется сложностью прохода: при низком балле за экзамен по математике вряд ли Вы пройдёте, несмотря на баллы по русскому, поэтому тут зависимость явно есть (при высоком проходном балле Вы проходите только тогда, когда у Вас высокие баллы за оба предмета).

Пункт в.

```python
math_not_ = np.array(math_not)
russian_not_ = np.array(russian_not)

corr_not = []
for i in range(len(math_not_)):
    corr_now = np.corrcoef(math_not_[i], russian_not_[i])
    corr_not.append(corr_now[0][1])
```

```python
plt.figure(figsize=(16, 10))

plt.plot(corr_not)
plt.xticks(np.arange(17), n_all)

plt.ylabel('Значение корреляции')
plt.xlabel('Порог')
plt.title('Зависимость между порогом и корреляцией между результатами за экзамен среди непрошедших')
plt.legend()
```

Здесь можно видеть обратную прошлой ситуацию: корреляция теперь стремится от -1 к 0.

Вспомним, что данный график показывает зависимость между баллами за два экзамена у непрошедших студентов.

Корреляция низкая при высоком проходном балле. Так происходит из-за того, что пройти на обучение при высоком пороге могут студенты с высокими баллами за два предмета, а остальные не проходят (т.е. попадают в выборку, корреляцию которой мы пытались построить на этом графике). Таким образом в эту выборку при высоком пороге попадают почти все комбинации баллов -> никакой корреляции здесь быть не может.

Также корреляция высокая при низком проходном балле - такая же логика только в обратную сторону: не проходят в универ при низком пороге только те, у кого и за русский, и за математику низкие баллы, а это означает, что тут будет корреляция.
